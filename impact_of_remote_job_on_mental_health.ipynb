##  Hypothesis Testing
*Test whether there is a significant difference in Mental Health Score between different Job Types.*



```python
from scipy.stats import f_oneway
from sklearn.preprocessing import LabelEncoder

# Convert 'Mental_Health_Condition' from string labels to numeric labels
le = LabelEncoder()
rj['Mental_Health_Condition_Encoded'] = le.fit_transform(rj['Mental_Health_Condition'])

# Perform ANOVA using the encoded mental health condition
job_groups = [rj[rj['Job_Role'] == group]['Mental_Health_Condition_Encoded'] for group in rj['Job_Role'].unique()]
stat, p_value = f_oneway(*job_groups)

# Print results
print(f'ANOVA test p-value: {p_value}')
```

## p-value = 0.0125
## This means that there is about a 1.25% chance that the differences in mental health conditions across the ## job roles occurred by random chance.
## Since 0.0125 < 0.05, we reject the null hypothesis.
